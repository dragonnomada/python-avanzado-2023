```python
import seaborn as sns
import numpy as np

edades = np.array([23, 54, 67, 34, 21, 32])

sns.distplot(edades)
```

```python
edades = np.array([23, 54, 67, 34, 21, 32, 43, 56, 32, 78, 12, 45, 36, 38, 72, 17])

sns.distplot(edades, bins=5, hist=True, kde=True, rug=True)
```

```python
muestras_tornillo = np.random.normal(4.5, 0.1, 100)

sns.distplot(muestras_tornillo)
```

```python
sns.boxplot(edades)
```

```python
sns.kdeplot(edades)
```

```python
sns.kdeplot(muestras_tornillo)
```

```python
sns.violinplot([edades, [6, 18, 24, 46, 56] ])
```

```python
# Suponer que tenemos 3 ejes de información

# 1er - Categoría (casado, divorciado, nunca-caso)
# 2da - Categoría (joven[20-24], no-tan-joven[24-30], adulto[31-36], más-adulto[36-45])
# 3er - Nivel estrés | Salario | Número hijos | Días de respuesta

mat = np.array(
    [
        # joven, nt-jove, adulto, m-adulto
        [0.8, 0.6, 0.73, 0.3], # casado
        [0.76, 0.45, 0.6, 0.23], # divorciado
        [0.45, 0.23, 0.67, 0.98], # no-casado
    ]
)

sns.heatmap(1 - mat)
```

```python
# Muestras de tornillos

# Clase Ancho Largo Diámetro Tipo
dataset = np.array(
    [
        [1, 8, 16, 4, 1], # 1
        [1, 6, 17, 4, 1], # 2
        [1, 7, 16, 5, 2], # 3
        [2, 14, 8, 4, 1], # 4
        [2, 15, 8, 6, 1], # 5
        [2, 14, 8, 6, 2], # 6
        [3, 12, 8, 9, 2], # 7
        [3, 8, 16, 10, 2], # 8
    ]
)

dataset
```

```python
import pandas as pd

sns.pairplot(pd.DataFrame(dataset, columns=["Clase", "Ancho", "Largo", "Diámetro", "Tipo"]))
```

```python
sns.pairplot(pd.DataFrame(dataset, columns=["Clase", "Ancho", "Largo", "Diámetro", "Tipo"]).query("Largo >= 16"))
```
